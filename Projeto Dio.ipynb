```python
from matplotlib import pyplot as plt
```

```python
# crecimento medio de acessos , por ano 
cm_acessos = df[['Acessos', 'Ano']].groupby('Ano').sum().reset_index()
plt.bar( cm_acessos['Ano'], cm_acessos['Acessos'])
```

```python
cm_acessos = df[['Acessos', 'UF']].groupby('UF').sum().reset_index()

plt.figure(figsize=(20,10))
plt.plot( cm_acessos['UF'], cm_acessos['Acessos'])
plt.xticks( rotation=45);
```

```python
cm_acessos = df[['Acessos', 'Mês']].groupby('Mês').sum().reset_index()

plt.figure(figsize=(20,10))
plt.plot( cm_acessos['Mês'], cm_acessos['Acessos'])
plt.xticks( rotation=45);
```
